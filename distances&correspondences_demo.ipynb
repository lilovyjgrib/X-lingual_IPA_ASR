```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
i = ct.Inventories()
```

```python
i.distance_matrix
```

```python
i.find_closest('ɔ', lang='english', k=3)
```

```python
i.find_closest('k͡p', lang='yoruba', k=3)
```

```python
examples = [
    ('ed͡ʒi', 'ɛd͡ʒi'),
    ('ki nĩ a mũ wa aje', 'kilæ mua jɪ')
]

for yo, eng in examples:
    alignment, score = ct.feature_edit_alignment(eng, yo)
    print(f"\n{eng} vs. {yo} — Distance: {round(score, 2)}")
    for src, tgt in alignment:
        print(f"{src:5} → {tgt:5}")

# Error rates
list_eng = ['kilæ mua jɪ', 'ɛd͡ʒi']
list_yor = ['ki nĩ a mũ wa aje', 'ed͡ʒi']

print(f"\nPhoneme error rate: {round(ct.phoneme_error_rate(list_eng, list_yor), 3)}")
```

```python
ct.make_confusion_matrix(list_eng, list_yor)
```

```python
data = ct.make_confusion_matrix(list_eng, list_yor).round(0)
colours = LinearSegmentedColormap.from_list("white_to_pink", ["#ffffff", "#ffc0cb"])

ill, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(data, annot=True, square=True, fmt=".0f", cmap=colours, ax=ax)

ax.set_xlabel("gold (yoruba) labels")
ax.set_ylabel("predicted (english) labels")
ax.set_title("example confusion matrix")
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
ax.xaxis.tick_top()
ax.tick_params(axis='x', top=True, length=0)

plt.tight_layout()
plt.show()
```

```python
### vibe-coded version
# Example matrix
df = ct.make_confusion_matrix(list_eng, list_yor).round(0)

# Custom colormap
colors = LinearSegmentedColormap.from_list("white_to_pink", ["#ffffff", "#ffc0cb"])

fig, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(
    df,
    annot=True,
    fmt=".0f",
    cmap=colors,
    square=True,
    linewidths=0.5,
    linecolor="#eeeeee",
    cbar_kws={"label": "%"},
    ax=ax
)

# Put x-axis labels on top
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')

# Remove tick marks ("sticks")
ax.tick_params(axis='x', top=True, length=0)
ax.tick_params(axis='y', length=0)  # optional: remove from y-axis too

# Set labels and title
ax.set_xlabel("Gold (Yoruba) Labels", fontsize=12, labelpad=10)
ax.set_ylabel("Predicted (English) Labels", fontsize=12, labelpad=10)
ax.set_title("Example Confusion Matrix", fontsize=14, fontweight='bold', pad=12)

# Format ticks
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="left", fontsize=10)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=10)

plt.tight_layout()
plt.show()
```
