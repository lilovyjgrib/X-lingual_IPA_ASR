```python
import zipfile
from pathlib import Path

zip_path = Path('/content/Data Sheet 1.zip')
extract_dir = Path('/content/data')

# Make sure the output directory exists
extract_dir.mkdir(parents=True, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zf:
    # Extract all files into extract_dir
    zf.extractall(extract_dir)

    # (Optional) list what was extracted
    for member in zf.namelist():
        print(f'Extracted: {member}')
```

```python
F_fname = 'data/Data_sent_publisher/Data/Formant_Poetry_Vowels.txt'
```

```python
import pandas as pd
import seaborn as sns

F_df = pd.read_csv(F_fname, sep='\t')
```

```python
F_df
```

```python
plot_F_df = F_df[(F_df['Filename'] == 'speaker1') & (F_df['Modes'] == 'Speech')]
```

```python
plot_F_df = plot_F_df[['Vowel', 'F1_50', 'F2_50']]
```

```python
IPA = {'E': 'ɛ', 'O': 'ɔ', 'a': 'a', 'i': 'i', 'u': 'u', 'o': 'o', 'e': 'e'}
plot_F_df['Vowel'] = plot_F_df['Vowel'].apply(lambda x: IPA[x])
```

```python
import matplotlib.pyplot as plt
```

```python
palette = sns.color_palette('husl', n_colors=plot_F_df.Vowel.nunique())

plt.figure(figsize=(6, 5))

for i, (vowel, grp) in enumerate(plot_F_df.groupby('Vowel')):
    sns.scatterplot(
        data=grp,
        x='F2_50', y='F1_50',
        color=palette[i],
        label=vowel,
        s=60,               # marker size
        alpha=0.6
    )
    sns.kdeplot(x=grp.F2_50, y=grp.F1_50,
                levels=[0.5],
                color=palette[i],
                linewidths=2,
                alpha=0.5,
                label=vowel)

    plt.legend(title='Vowel', frameon=True, loc='upper right')

# Phonetic convention: F1 ↑ downwards, F2 ↑ leftwards
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()

plt.xlabel('F2 (Hz)')
plt.ylabel('F1 (Hz)')
plt.title('Vowel Space')
plt.tight_layout()
plt.show()
```

```python
F_df = pd.read_csv('TIMIT_vowels_FCJFO', sep='\t')
```

```python
arpa_to_ipa = {
    "aa": "ɑ",    # father
    "ae": "æ",    # trap
    "ah": "ʌ",    # strut
    "ao": "ɔ",    # thought
    "aw": "aʊ",   # mouth
    "ay": "aɪ",   # price
    "ax": "ə",    # comma (schwa)
    "ax-h": "ə",  # aspirated schwa (treat as schwa)
    "axr": "ɚ",   # nurse (r-colored schwa)
    "eh": "ɛ",    # dress
    "er": "ɝ",    # bird (r-colored vowel)
    "ey": "eɪ",   # face
    "ih": "ɪ",    # kit
    "ix": "ɨ",    # high central unrounded (near-schwa)
    "iy": "i",    # fleece
    "ow": "oʊ",   # goat
    "oy": "ɔɪ",   # choice
    "uh": "ʊ",    # foot
    "uw": "u",    # goose
    "ux": "ʉ"     # dude
}

diphthongs = {"aw", "ay", "ey", "ow", "oy"}
```

```python
is_diph = F_df.Vowel.apply(lambda x: x in diphthongs)
```

```python
F_df_diph = F_df[is_diph]
```

```python
F_df_monoph = F_df[~is_diph]
```

```python
F_df_monoph = F_df_monoph[['Vowel', 'F1_50', 'F2_50']]
F_df_monoph.columns = ['Vowel', 'F1', 'F2']
```

```python
F_df_diph = F_df_diph[['Vowel', 'F1_20', 'F2_20']]
F_df_diph.columns = ['Vowel', 'F1', 'F2']
```

```python
plot_F_df = pd.concat([F_df_monoph, F_df_diph], ignore_index=True)
```

```python
plot_F_df['Vowel'] = plot_F_df['Vowel'].apply(lambda x: arpa_to_ipa[x])
```

```python
import numpy as np
is_freq = plot_F_df.Vowel.apply(lambda x: plot_F_df.Vowel.value_counts()[x] > 3)
plot_F_df = plot_F_df[is_freq]
```

```python
palette = sns.color_palette('husl', n_colors=plot_F_df.Vowel.nunique())

plt.figure(figsize=(6, 5))

for i, (vowel, grp) in enumerate(plot_F_df.groupby('Vowel')):
    sns.scatterplot(
        data=grp,
        x='F2', y='F1',
        color=palette[i],
        label=vowel,
        s=60,               # marker size
        alpha=0.6
    )
    sns.kdeplot(x=grp.F2, y=grp.F1,
                levels=[0.5],
                color=palette[i],
                linewidths=2,
                alpha=0.5,
                label=vowel)

    plt.legend(title='Vowel', frameon=True, loc='upper right')

# Phonetic convention: F1 ↑ downwards, F2 ↑ leftwards
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()

plt.xlabel('F2 (Hz)')
plt.ylabel('F1 (Hz)')
plt.title('Vowel Space')
plt.tight_layout()
plt.show()
```
